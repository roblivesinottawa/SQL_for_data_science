ROLLUP(): 

```python
query = """
select
   cr.country_name, cr.region_name, count(e.*)
from
   employees e
join
   company_regions cr
on
   e.region_id = cr.id
group by
   rollup(cr.country_name, cr.region_name)
"""
engine.execute(query).fetchall()
```

CUBE():

```python
query = """
    SELECT
        cr.country_name,
        cr.region_name,
        cd.department_name,
        count(e.*)
    FROM
        employees e
    JOIN
        company_regions cr
    ON
        e.region_id = cr.id
    JOIN
        company_departments cd
    ON 
        e.department_id = cd.id
    GROUP BY
        CUBE(1, 2, 3)
    ORDER BY
        1, 2, 3;
"""
engine.execute(query).fetchall()
```

WINDOW FUNCTIONS AND ORDERED DATA

```python
query = """
select
   department_id,
   last_name,
   salary,
   first_value(salary) over -- function that returns the first value of the window
        (partition by 
            department_id 
        order by 
            salary asc) first_sal 
from
    employees;

"""

engine.execute(query).fetchall()
```

nth_value and ntile

```python
query = """
    select
        department_id,
        salary,
        avg(salary) over (partition by department_id)
    from
        employees;
"""

engine.execute(query).fetchall()
```

```python
query = """
    select
        department_id,
        salary,
        ntile(4) over ( -- ntile(#) returns the nth tile of the window
            partition by department_id 
            order by salary desc) as quartile
    from
    employees;
"""
engine.execute(query).fetchall()
```

COMMON TABLE EXPRESSIONS

- auxiliary statements for use in larger queries 
- create temporary tables used only within a query 
- used to simplify complex queries that would otherwise require subqueries

```python
# get some information on salaries across regions
# and list all of the regions with above average salaries

query = """
    WITH region_salaries AS
 (SELECT
    e.region_id,
    SUM(e.salary) region_salary
  FROM
    employees e
  GROUP BY 
    region_id),
 top_region_salaries AS
  (SELECT
      region_id
   FROM
      region_salaries
    WHERE region_salary > (SELECT SUM(region_salary)/7 FROM region_salaries))
    SELECT *
    FROM region_salaries
    WHERE
        region_id IN (SELECT region_id FROM top_region_salaries);
"""
engine.execute(query).fetchall()
```

```python
query = """
    with recursive report_structure(id, department_name,parent_department_id) as
    (select id,  department_name, parent_department_id 
    from org_structure where id = 7
    union
    select os.id, os.department_name, os.parent_department_id 
    from org_structure os
    join report_structure rs on rs.parent_department_id = os.id
    )
    select
    *
    from report_structure;
"""
engine.execute(query).fetchall()
```
